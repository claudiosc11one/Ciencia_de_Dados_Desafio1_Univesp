```python
import pandas as pd
import seaborn as sns
```

```python
uri="https://raw.githubusercontent.com/fivethirtyeight/data/master/airline-safety/airline-safety.csv"
seguranca_aerea=pd.read_csv(uri, error_bad_lines=False, encoding='latin-1')
seguranca_aerea.head()
```

```python
seguraca_aerea.columns = ["cia_aerea", "lugar_Disponivel(km/semana)", "incidentes_85_99", "acidentes_fatais_85_99", "fatalidades_85_99", "incidentes_00_14", "acidentes_fatais_00_14", "fatalidades_00_14"]
seguraca_aerea.head()
```

```python
seguraca_aerea["incidentes_85_99"].mean()
```
